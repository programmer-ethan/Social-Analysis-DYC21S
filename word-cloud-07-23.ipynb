```python
# wordcloud -> 단어 말풍선 >pip install wordcloud
from wordcloud import WordCloud, STOPWORDS
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
text=open("09. a_new_hope.txt").read()
text=text.replace("Luke's","Luke")
text=text.replace("LUKE's","Luke")
text=text.replace("HAN","Han")
storm_mask=np.array(Image.open('09. stormtrooper_mask.png'))
stopwords = set(STOPWORDS)

stopwords.add("one")
stopwords.add("two")
stopwords.add("look")
stopwords.add("move")
stopwords.add("toward")
stopwords.add("back")
wc=WordCloud(background_color='black', max_words=2000, mask=storm_mask, stopwords=stopwords)
wc=wc.generate(text)
wc.words_
```

```python
plt.figure(figsize=(15,15))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import random
def grey_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    return 'hsl(0,0%%,%d%%)' % random.randint(60,100)
```

```python
plt.figure(figsize=(15,15))
plt.imshow(wc.recolor(color_func=grey_color_func))
plt.axis('off')
plt.show()
```

```python
from konlpy.tag import Kkma
from konlpy.corpus import kolaw
from konlpy.utils import pprint
from nltk import collocations

measures = collocations.BigramAssocMeasures()
tagged_words = Kkma().pos(doc_ko)
```

```python
tagged_words
```

```python
words = [w for w,t in tagged_words]
stop_words = [u'에서']
words
```

```python
finder = collocations.BigramCollocationFinder.from_words(words)
finder.apply_word_filter(lambda w: len(w) < 2 or w in stop_words)
finder.apply_freq_filter(5)
pprint(finder.nbest(measures.pmi,20))
```
